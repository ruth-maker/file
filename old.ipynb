```python
movie_info = pd.read_csv('rt.movie_info.tsv',error_bad_lines=False,sep='\t')
```

```python
# Getting first 200 movies 
first_200 = []
for i in movie_info.genre[:202].dropna():
    first_200.append(i)   
first_200
```

```python
# putting movies into groups
grouped_200 = []
for i in first_200:
    grouped_200.append(i.split('|',1)[0])
grouped_200=collections.Counter(grouped_200)
grouped_200
```

```python
# Getting the nmeas of the first 200 movies in the Drama genre
movie_buddget = pd.read_csv('tn.movie_budgets.csv')
name_movies = []
for x,y in enumerate(first_200):
    if y[0]=='D': name_movies.append(movie_buddget.movie[x])
name_movies
```

```python
# Getting product budget of the first 200 movies in Drama genre
product_budget = []
for x,y in enumerate(first_200):
    if y[0]=='D': product_budget.append(movie_buddget.production_budget[x])
product_budget
```

```python
# Getting ratings of the first 200 movies in Drama genre
ratings = pd.read_csv('title.ratings.csv')
rating = []
for x,y in enumerate(first_200):
    if y[0]=='D': rating.append(ratings.averagerating[x])
rating
```

```python
# Getting system ratings of the first 200 movies in Drama genre
system_rating = []
for x,y in enumerate(first_200):
    if y[0]=='D': system_rating.append(movie_info.rating[x])
system_rating
```

```python
# Getting domestic gross of the first 200 movies in Drama genre
domestic_gross = []
for x,y in enumerate(first_200):
    if y[0]=='D': domestic_gross.append(movie_buddget.domestic_gross[x])
domestic_gross
```

```python
# Getting worldwide gross of the first 200 movies in Drama genre
worldwide_gross = []
for x,y in enumerate(first_200):
    if y[0]=='D': worldwide_gross.append(movie_buddget.worldwide_gross[x])
worldwide_gross
```

```python
all_drama_info = pd.DataFrame({'Name':name_movies,'System_ratings':system_rating,'Rating':rating,'Worlwide_gross':worldwide_gross,
                               'Domestic_gross':domestic_gross,'Product_budget':product_budget})
all_drama_info1=all_drama_info.dropna()
all_drama_info1 = all_drama_info[all_drama_info.Name !='Moonfall']
all_drama_info1
```

```python
# turning Produt budget gross into int
storage1 = []
storage2 = []
budget=[]
for i in  all_drama_info1.dropna().Product_budget:
    storage1.append(i.replace('$',''))
for i in storage1:
    storage2.append(i.replace(',',''))
for i in storage2:   
    i = int(i)
    budget.append(i)
len(budget)
```
